```python
# pos

from pycorenlp import StanfordCoreNLP
nlp = StanfordCoreNLP('http://localhost:9000')

text = ('M and N walked towards the city. They did not want to walk back')

output = nlp.annotate(text, properties={
    'annotators': 'pos',
    'outputFormat': 'json'
})

print(output)
```
